```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_Toko_Online.csv")
print(df.head())
```

```python
print(df.info())
print(df.isnull().sum())
df = df.dropna()
```

```python
print("Statistik Deskriptif:")
print(df.describe())

print("\nProduk Terlaris:")
print(df.groupby("Produk")["Jumlah"].sum().sort_values(ascending=False))
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Produk", y="Jumlah")
plt.title("Total Pendapatan per Produk")
plt.show()
plt.figure(figsize=(30,10))
sns.lineplot(data=df, x="Tanggal", y="Pendapatan", marker="o")
plt.title("Tren Pendapatan Mingguan")
plt.show()
```

```python
df.to_csv("hasil_analisis_penjualan.csv", index=False)
print("Data hasil analisis telah disimpan sebagai Data_Toko_Online.csv")
```

```python
df['Diskon'] = df['Harga'] * 0.10
```

```python
df['Pendapatan Bersih'] = df['Pendapatan'] - df['Diskon']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Produk", y="Pendapatan Bersih", estimator=sum)
plt.show()
```

```python
df.to_csv('analisis_diskon.csv', index=False)
```
